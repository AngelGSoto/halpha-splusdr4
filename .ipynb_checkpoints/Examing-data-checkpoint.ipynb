```python
# See the dimesion of the CCD
print(tab['X_IMAGE'].max())
print(tab['Y_IMAGE'].max())
```

```python
# Masking, usar FLATS menor que 4
# Masking
m_x = (tab["R_APER_3"] > 16) & (tab["R_APER_3"] <= 17.5) & (tab["CLASS_STAR_R"] >= 0.6)
m_err = (tab["e_R_APER_3"]  <= 0.2) & (tab["e_F660_APER_3"]  <= 0.2) & (tab["e_I_APER_3"]  <= 0.2)
m_sn = (tab["SNR_WIN_R"]  >= 15) & (tab["SNR_WIN_F660"]  >= 15) & (tab["SNR_WIN_I"]  >= 15)

valid_ellipticity_range = (0.1, 0.8)  # Example valid range for ellipticity
valid_elongation_range = (1.0, 3.0)   # Example valid range for elongation
shape_mask = (tab['ELLIPTICITY'] >= valid_ellipticity_range[0]) & \
             (tab['ELLIPTICITY'] <= valid_ellipticity_range[1]) & \
             (tab['ELONGATION'] >= valid_elongation_range[0]) & \
             (tab['ELONGATION'] <= valid_elongation_range[1])

valid_fwhm_range = (1.0, 3.0)  # Example valid range for FWHM in pixels
fwhm_mask = (tab['FWHM_IMAGE_R'] >= valid_fwhm_range[0]) & \
            (tab['FWHM_IMAGE_R'] <= valid_fwhm_range[1])

mask = m_x & m_err & m_sn & shape_mask & fwhm_mask  # Use fwhm_mask instead of valid_fwhm_range
tab_clean = tab[mask]
```

```python
# Define the dimensions of your CCD
ccd_width = int(tab_clean['X_IMAGE'].max())  # Replace with the actual width of your CCD image
ccd_height = int(tab_clean['Y_IMAGE'].max())  # Replace with the actual height of your CCD image

# Define the buffer zone size for corner removal
buffer_size = 50  # Adjust as needed

# Define the corners of the CCD
corners = [(0, 0), (0, ccd_height), (ccd_width, 0), (ccd_width, ccd_height)]

# Create a mask to identify objects in the corners
corner_mask = np.zeros(len(tab_clean), dtype=bool)

for corner in corners:
    corner_x, corner_y = corner
    corner_distance = np.sqrt((tab_clean['X_IMAGE'] - corner_x)**2 + (tab_clean['Y_IMAGE'] - corner_y)**2)
    corner_mask |= corner_distance < buffer_size

# Apply the corner mask to filter out objects in the corners
tab_clean_filtered = tab_clean[~corner_mask]
```

```python
ri = tab_clean["R_APER_3"] - tab_clean["I_APER_3"]
rj660 = tab_clean["R_APER_3"] - tab_clean["F660_APER_3"]
```

```python
def equation(x):
    # Define your equation here, for example, a linear equation: y = mx + c
    m = 0.35  # Slope
    c = -0.35  # Intercept
    return m * x + c
```

```python
def plots(color1, color2, x_values, y_values, xlim=None, ylim=None):
    # Calculate the point density
    x, y = color1, color2
    xy = np.vstack([x, y])
    z = gaussian_kde(xy)(xy)
    # Sort the points by density, so that the densest points are plotted last
    idx = z.argsort()
    x, y, z = x[idx], y[idx], z[idx]
    fig, ax = plt.subplots(figsize=(14, 11))
    ax_ = ax.scatter(x, y, c=z, s=10)
    plt.colorbar(ax_)
    # Create line
    plt.plot(x_values, y_values)
    # Set x and y limits
    if xlim is not None:
        plt.xlim(xlim)
    if ylim is not None:
        plt.ylim(ylim)
    plt.show()
```

```python
# Generate x values
x_values = np.linspace(-10, 10, 100)  # Generate 100 x values from 0 to 10

# Compute y values using the equation
y_values = equation(x_values)
```

```python
plots(ri, rj660, x_values, y_values, xlim=(-1.5, 3), ylim=(-5, 4))
```

```python
# Select objects below the line
# below_line = rj660 < equation(ri)
color_crit = rj660  <= 0.35*ri - 0.35
```

```python
tab_clean_below = tab_clean[color_crit]
```

```python
plots(tab_clean_below["R_APER_3"] - tab_clean_below["I_APER_3"], tab_clean_below["R_APER_3"] - tab_clean_below["F660_APER_3"], x_values, y_values, xlim=(-1.5, 3), ylim=(-5, 4))
```
