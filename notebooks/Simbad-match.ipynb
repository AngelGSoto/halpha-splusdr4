```python
dict_simbad = {}
for i in np.unique(df["main_type"]):
    m = df["main_type"] == i
    type_simbad = df[m]
    dict_simbad[i] = len(type_simbad)
```

```python
dict_simbad
```

```python
# Count the number of objects of each type and store it in a dictionary
dict_simbad_ = df["main_type"].value_counts().to_dict()
```

```python
dict_simbad_
```

```python
# Grouping and summing the counts by main type for the main survey
grouped_counts_main = {
    'Stellar Binary System': sum([dict_simbad_[key] for key in ['CataclyV*', 'CV*_Candidate', 'RSCVn', 'EB*', 'EB*_Candidate', 'SB*_Candidate']]),
    'Variable Star': sum([dict_simbad_[key] for key in ['PulsV*', 'V*', 'PulsV*delSct', 'RotV*', 'RRLyr']]),
    'Star': sum([dict_simbad_[key] for key in ['Star', 'Blue', 'low-mass*', 'WD*', 'WD*_Candidate', 'PM*', 'BlueStraggler']]),
    'Radio Source': sum([dict_simbad_[key] for key in ['Radio', 'Radio(cm)', 'RadioG']]),
    'Active Galactic Nucleus (AGN)': sum([dict_simbad_[key] for key in ['AGN', 'AGN_Candidate', 'Seyfert_1']]),
    'Quasar': sum([dict_simbad_[key] for key in ['QSO', 'QSO_Candidate']]),
    'Galaxy': dict_simbad_['Galaxy'],
    'Other': sum([dict_simbad_[key] for key in ['Hsd_Candidate', 'Pec*', 'AGB*', 'MIR']])
}
```

```python
# Calculating the total count
total_main = sum(grouped_counts_main.values())
grouped_counts_main, total_main
```

```python
# Count the number of objects of each type and store it in a dictionary
dict_simbad_disk = df_disk["main_type"].value_counts().to_dict()
```

```python
dict_simbad_disk
```

```python
# Grouping and summing the counts by main type
grouped_counts_disk = {
    'Emission-line star': dict_simbad_disk['Em*'] + dict_simbad_disk['Be*'],
    'Young stellar object': sum([dict_simbad_disk[key] for key in ['YSO_Candidate', 'YSO', 'Orion_V*', 'TTau*_Candidate']]),
    'Stellar Binary System': sum([dict_simbad_disk[key] for key in ['CataclyV*', 'CV*_Candidate', 'RSCVn', 'EB*', 'EB*_Candidate', 'SB*']]),
    'Variable star': sum([dict_simbad_disk[key] for key in ['PulsV*delSct', 'PulsV*', 'LPV*', 'LP*_Candidate', 'Mira', 'RRLyr', 'V*', 'V*?_Candidate', 'BYDra']]),
    'Star': sum([dict_simbad_disk[key] for key in ['Star', '**', 'RGB*', 'C*', 'WD*_Candidate']]),
    'Nebula': sum([dict_simbad_disk[key] for key in ['PN?_Candidate', 'RfNeb', 'Nova']]),
    'Other': sum([dict_simbad_disk[key] for key in ['EmObj', 'Hsd_Candidate', 'deltaCep', 'Cepheid_Candidate', 'Transient', 'X']])
}
```

```python
# Calculating the total count
total_disk = sum(grouped_counts_disk.values())
grouped_counts_disk, total_disk
```
