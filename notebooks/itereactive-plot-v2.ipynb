```python
%matplotlib widget
import matplotlib.pyplot as plt
import numpy as np
from IPython.display import display
import ipywidgets as widgets

# Crear un widget de texto
text_area = widgets.Textarea(
    value='',
    placeholder='Pendientes e interceptos de las lÃ­neas dibujadas aparecerÃ¡n aquÃ­...',
    description='Resultados:',
    layout=widgets.Layout(width='100%', height='200px'),
    disabled=False
)
display(text_area)

# Inicializa la figura y el eje
fig, ax = plt.subplots()
ax.set_xlim([-10, 10])
ax.set_ylim([-10, 10])
ax.grid(True)
points = []
line_count = 0

# FunciÃ³n para calcular la pendiente y la intersecciÃ³n
def calculate_slope_intercept(x1, y1, x2, y2):
    if x1 == x2:
        m = np.inf
        b = np.nan
    else:
        m = (y2 - y1) / (x2 - x1)
        b = y1 - m * x1
    return m, b

# FunciÃ³n de evento para manejar clics
def on_click(event):
    global line_count
    if event.inaxes is not None:
        points.append((event.xdata, event.ydata))
        if len(points) == 2:
            x1, y1 = points[0]
            x2, y2 = points[1]
            m, b = calculate_slope_intercept(x1, y1, x2, y2)
            line_count += 1
            line_label = f'Linea{line_count}'
            # Dibuja la lÃ­nea
            ax.plot([x1, x2], [y1, y2], marker='o')
            # Etiqueta de la lÃ­nea en el grÃ¡fico
            ax.text((x1 + x2) / 2, (y1 + y2) / 2, line_label, fontsize=12, color='red')
            fig.canvas.draw()
            # Actualizar el cuadro de texto con la pendiente y la intersecciÃ³n
            text_area.value += f'{line_label} = Pendiente: {m:.6f}, Intercepto: {b:.6f}\n'
            points.clear()

# Conectar el evento de clic al grÃ¡fico
cid = fig.canvas.mpl_connect('button_press_event', on_click)

plt.show()

# Verificar que el evento estÃ¡ conectado
print(f"Event connection id: {cid}")
```
