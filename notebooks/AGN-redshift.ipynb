```python
from astropy.io import fits
import numpy as np
from astropy.table import Table
import pandas as pd
import matplotlib.pyplot as plt
import glob
from astropy.table import vstack
import seaborn as sns
```

```python
# Lista de archivos CSV
file_list = ['../Class_main_unique/Halpha_emitter_group0-simbad.csv', '../Class_main_unique/Halpha_emitter_group1-simbad.csv', '../Class_main_unique/Halpha_emitter_group2-simbad.csv', '../Class_main_unique/Halpha_emitter_group3-simbad.csv', '../Class_main_unique/Halpha_emitter_group4-simbad.csv']  # Reemplaza con tus archivos
```

```python
# Crear una figura para las subgráficas
plt.figure(figsize=(12, 20))

# Iterar sobre los archivos y crear una subgráfica para cada uno
for i, file in enumerate(file_list):
    # Cargar datos del archivo
    df = pd.read_csv(file)
    
    # Graficar el histograma para cada archivo
    plt.subplot(len(file_list), 1, i + 1)
    counts, bin_edges, _ = plt.hist(df["redshift"], bins=100, color='skyblue', edgecolor='black')
    
    # Estimar el pico de la distribución
    peak_index = np.argmax(counts)
    peak_redshift = bin_edges[peak_index]
    plt.axvline(x=peak_redshift, color='r', linestyle='--', label=f'Pico: {peak_redshift:.2f}')
    
    # Añadir título y etiquetas
    plt.title(f'Distribución de Redshift para {file}')
    plt.xlabel('Redshift')
    plt.ylabel('Frecuencia')
    plt.grid(True)  # Añadir líneas de la cuadrícula para mejor legibilidad
    plt.legend()

# Ajustar el layout para evitar solapamientos
plt.tight_layout()

# Mostrar la figura con todas las subgráficas
plt.show()
```
